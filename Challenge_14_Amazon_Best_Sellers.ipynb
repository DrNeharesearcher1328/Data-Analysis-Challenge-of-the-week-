# 3. Are there any spelling errors in the Name column?

```python
pip install fuzzywuzzy
```

```python
import pandas as pd
from fuzzywuzzy import fuzz, process

# Load the dataset
Bestsellers = pd.read_csv('/content/bestsellers with categories.csv')

# Set threshold for matching
threshold = 90

# Identify potential duplicates using fuzzy matching
duplicates = []
for column in ['Name', 'Author']:
    unique_values = Bestsellers[column].unique()
    for value in unique_values:
        matches = process.extractBests(value, unique_values, scorer=fuzz.token_set_ratio, score_cutoff=threshold)
        for match, score in matches:
            if match != value and {value, match} not in duplicates:
                duplicates.append((value, match, score))

# Print the potential duplicates
print("Potential duplicates:")
for value1, value2, score in duplicates:
    print(f"Value 1: {value1}\tValue 2: {value2}\tScore: {score}")

# Replace spelling errors in the dataset
replace_values = {
    'George R. R. Martin': 'George R.R. Martin',
    'J. K. Rowling': 'J.K. Rowling',
    'The Hunger Games': 'The Hunger Games (Book 1)',
    'Eclipse (Twilight)': 'Eclipse (Twilight Sagas)',
    'The Girl Who Played with Fire (Millennium)': 'The Girl Who Played with Fire (Millennium Series)'
}
Bestsellers.replace(replace_values, inplace=True)

# Print the modified dataset
print("\nModified dataset:")
print(Bestsellers)
```

# 4.Create a plot showing the top 10 authors by rating.

```python
import pandas as pd
import plotly.express as px

# Assuming the dataset is already loaded into a DataFrame called 'bestsellers'

top_10_authors = Bestsellers.groupby("Author")["User Rating"].mean().nlargest(10).reset_index()

fig = px.bar(top_10_authors, x="User Rating", y="Author", orientation="h", title="Top 10 Authors by Rating",
             color="Author", color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_layout(xaxis_title="Average Rating", yaxis_title="Author")
fig.show()
```

# 6.Create a plot showing the top 5 books by number of reviews.

# 1.   What are the average ratings for each genre of books?




```python
# What are the average ratings for each genre of books?
import matplotlib.pyplot as plt

# Calculate average ratings by genre
average_ratings = Bestsellers.groupby('Genre')['User Rating'].mean()

# Create a bar plot to visualize the average ratings
plt.figure(figsize=(8, 6))
average_ratings.plot(kind='bar', color='blue')
plt.title('Average Ratings by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```



--- **Observation:-** The average ratings for Non-Fiction books are slightly higher than those for Fiction books, but the difference is not significant. Both genres receive generally positive ratings, with an average rating of around 4.6. This suggests that readers enjoy and recommend books from both genres.


# 2. How does the publication year affect the average user rating for the bestsellers in the dataset?



```python
import matplotlib.pyplot as plt

# Group the data by publication year and calculate the average user rating for each year
average_ratings_by_year = Bestsellers.groupby('Year')['User Rating'].mean()

# Create a line plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.plot(average_ratings_by_year.index, average_ratings_by_year.values, marker='o')
plt.title('Average User Rating by Publication Year')
plt.xlabel('Year')
plt.ylabel('Average User Rating')
plt.xticks(rotation=45)
plt.show()
```



--- **Observation:-** The line plot depicting the average user rating by publication year shows the trend in reader ratings over time. The plot indicates that there is some variation in the average user ratings across different years. The ratings fluctuate between approximately 4.5 and 4.7, with some years showing higher average ratings compared to others. This suggests that reader perceptions and preferences may have evolved over time, resulting in variations in the ratings given to books published in different years. However, it is important to note that the observed differences are relatively small, indicating that overall, readers have consistently enjoyed the books across different publication years.

# 9. What is the average rating for books priced above the specified threshold, and how does it compare to the overall average rating?

```python
import seaborn as sns
import matplotlib.pyplot as plt

threshold = 15  # Set the price threshold

# Filter the dataset for books priced above the threshold
above_threshold = Bestsellers[Bestsellers['Price'] > threshold]

# Calculate the average rating for books above the threshold
average_rating_above_threshold = above_threshold['User Rating'].mean()

# Create a bar plot to visualize the average rating
plt.figure(figsize=(6, 4))
sns.barplot(x=['Above Threshold'], y=[average_rating_above_threshold])
plt.title('Average Rating for Books Priced Above Threshold')
plt.xlabel('Price Threshold')
plt.ylabel('Average Rating')
plt.show()
```



**Observation:-**The bar plot represents the average rating for books priced
above a threshold of 15. The plot shows that books priced above the threshold have an average rating of approximately 4.57. This suggests that higher-priced books tend to receive positive ratings from readers.
